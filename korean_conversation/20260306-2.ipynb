```python
from model import Transformer
import tensorflow_datasets as tfds
```

```python
tokenizer = tfds.deprecated.text.SubwordTextEncoder.load_from_file("./tokenizer")
```

```python
VOCAB_SIZE = 8096
NUM_LAYERS = 2
D_MODEL = 256
NUM_HEADS = 8
DFF = 512
DROPOUT = 0.1
model = Transformer(
        vocab_size=VOCAB_SIZE,
        num_layers=NUM_LAYERS,
        dff=DFF,
        d_model=D_MODEL,
        num_heads=NUM_HEADS,
        dropout=DROPOUT
    ).to('cpu')
#mac은 mps
```

```python
import torch

model_dict = torch.load("./ko-con.pth", map_location='cpu')

# 전이 학습(Transfer Learning)
# 학습 가능한 파라미터가 있는 레이어(Conv, Linear 등)만 키(Key)값으로 가져오기
model.load_state_dict(model_dict) 
```
