## MongoDB에서 RAG 데이터 불러오기

```python
# MongoDB 연결 정보 (저장할 때와 동일하게 유지)
from pymongo import MongoClient

MONGO_URI = "mongodb+srv://cudabpy_db_user:v4tMuaMWPJ17pL5L@korean-conversation.0cmrkx4.mongodb.net/?appName=Korean-Conversation"
DB_NAME = "korean_conversation"
TARGET_COLLECTION = "rag_data"

print("🌐 MongoDB에서 RAG 데이터를 불러오는 중...")
try:
    client = MongoClient(MONGO_URI)
    collection = client[DB_NAME][TARGET_COLLECTION]
    # '_id'는 제외하고 순수 데이터만 리스트 형태로 가져옵니다.
    rag_dict_list = list(collection.find({}, {'_id': 0}))
    client.close()
    print(f"✅ 총 {len(rag_dict_list)}개의 데이터를 성공적으로 불러왔습니다!")
except Exception as e:
    print(f"❌ MongoDB 연결 또는 데이터 로드 실패: {e}")
```

## LangChain Document 객체로 변환, 텍스트 스플릿

```python
# LangChain Document 객체로 변환
from langchain_core.documents import Document
from langchain_text_splitters import RecursiveCharacterTextSplitter

print("📝 데이터를 LangChain 형식(Document)으로 변환 중...")
docs = []
for item in rag_dict_list:
    # RAG 검색이 문맥을 잘 파악할 수 있도록 Q와 A를 하나로 묶어줍니다.
    text_content = f"Q: {item.get('question', '')}\nA: {item.get('answer', '')}"
    
    # 메타데이터에 원본 번호나 출처를 남겨두면 나중에 출처 표기할 때 매우 유용합니다.
    metadata = {"source": "mongodb_rag", "no": str(item.get('no', ''))}
    doc = Document(page_content=text_content, metadata=metadata)
    docs.append(doc)


# 텍스트 스플릿 (문서 쪼개기)
# 대화가 너무 길면 모델이 헷갈릴 수 있으므로 적절한 길이로 잘라줍니다.
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
splits = text_splitter.split_documents(docs)
print(f"✂️ 텍스트 분할 완료: 총 {len(splits)}개의 조각(Chunk)이 생성되었습니다.")
```

## 임베딩 모델 로드 및 Chroma DB 영구 저장 (핵심)

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma

print("🧠 임베딩 모델 로드 및 Chroma DB 구축을 시작합니다. (데이터 양에 따라 수 분이 소요될 수 있습니다...)")

# 한국어 특화 임베딩 모델
embeddings = HuggingFaceEmbeddings(model_name="jhgan/ko-sroberta-multitask")

# 벡터 DB 생성 및 로컬 디렉토리에 저장
# persist_directory="./rag_db" 설정 덕분에 현재 폴더에 rag_db 폴더가 생깁니다.
vectorstore = Chroma.from_documents(
    documents=splits,
    embedding=embeddings,
    persist_directory="./rag_db", 
    collection_metadata={"hnsw:space": "cosine"} # 유사도 측정 방식을 코사인 유사도로 설정
)

print("🎉 완벽합니다! 작업 중인 폴더에 'rag_db' 폴더가 성공적으로 생성되었습니다.")
```
